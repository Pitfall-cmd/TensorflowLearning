多变量房价预测

```python
import matplotlib.pyplot as plt
from mpl_toolkits import  mplot3d
df1=pd.read_csv('data1.csv',names=['square','bedrooms','price'])
df1.head()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.set_xlabel('square')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')
ax.scatter3D(df1['square'],df1['bedrooms'],df1['price'],c=df1['price'],cmap='Reds')
```

## 数据规范化

```python
def normalize_feature(df):
    return df.apply(lambda column:(column-column.mean())/column.std())

df=normalize_feature(df1)
df.head()
```

```python
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.set_xlabel('square')
ax.set_ylabel('bedrooms')
ax.set_zlabel('price')
ax.scatter3D(df['square'],df['bedrooms'],df['price'],c=df1['price'],cmap='Reds')
```
